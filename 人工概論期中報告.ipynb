```python
#用count函數進行數值累加
sum = 0
for count in range(1, 11, 1): #從1開始，間隔1, 不包含11
    sum += count
print("1+2+3+...+10 =", sum)
```

```python
#range另外一種寫法
sum = 0
for count in range(11): #1-10, 不包含11 ()內的數值決定最大極限
    sum += count
print("1+2+3+...+10 =", sum)
```

```python
#實際運用 if else range
#雞兔同籠，若共有35個頭，94隻腳，請問有幾隻雞?幾隻兔?
for chicken in range(0,36):
    rabbit=35-chicken
    if (2*chicken+4*rabbit) == 94:
        print("雞有",chicken,"隻,兔子有",rabbit,"隻")
        break
else:
    print("無解")
```

```python
#進行while迴圈 在條件符合下不斷執行程式碼
sum=0
count=1
while count <=10:
    sum += count
    count += 1
print("1+2+3+...+10=",sum)
```

```python
#實際運用 帳號密碼判斷
id="tony"
pwd="1234"
while True: #將條件設定x和y皆為Ture
    x=input("請輸入帳號：")
    y=input("請輸入密碼：")
    if (x==id and y==pwd): #判斷輸入是否和預設值相同
        print("歡迎登入")
        break
    else:
        print("帳號或密碼輸入錯誤")
```

```python
#random實際運用 猜數字遊戲
import random
number=random.randint(1,20) #範圍設定在1至20之間
while True:
    guess=int(input("請輸入1~20之間的數字-->"))
    if guess == number: #判斷是否符合
        print("你猜對了，數字是", number)
        break
    else:
        print("你猜錯了")
```
